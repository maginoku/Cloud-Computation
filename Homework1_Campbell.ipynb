```python
# Creating a function to make everything lowercase and special characters

def file_clean(x):
    x = x.lower()
    x = re.sub("[^a-z ]+", '', x)
    return x
```

```python
# here we will filter and map so we can fulfill the tasks that was assigned

Results = (
    Input.map(lambda x: file_clean(x))
    .filter(lambda x: x!="")
    .map(lambda x: x.split(" "))
    .flatMap(lambda x: itertools.permutations(x,2))
    .filter(lambda x: x[0] !="")
    .filter(lambda x: x[1] !="")
    .map(lambda x: (x,1))
    .reduceByKey(lambda x, y: x + y)
    .collect()
)
```
